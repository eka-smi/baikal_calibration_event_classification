```python
#время прохождения света от лазера до ОМ, находящегося на той же высоте
file2=open(file_path2,'r')
lines2=file2.readlines()
lines2=[line.split() for line in lines2]
lines2
columns2=['N','X','Y', 'Z']
df2 = pd.DataFrame(lines2, columns=columns2)
df2['N'] = df2['N'].astype(int)
df2[['X','Y', 'Z']] = df2[['X','Y', 'Z']].astype(float)
laser_coords=np.array([9830.0,-1100.0,6890.0])

def distance(row):
    point = np.array([row['X'], row['Y'],row['Z']])
    return np.linalg.norm(point-laser_coords)

df2['Distance'] = df2.apply(distance, axis=1)
df2['Time']= df2.apply(distance, axis=1)*4.57/100

#гистограмма для времен дохождения света от лазера до ОМ
y=df2['Distance']
plt.hist(y, bins=np.linspace(0, 50000, 588), edgecolor='black')
plt.show()
df2['Time'].max()-df2['Time'].min()
```

```python
# time_diff = df.groupby('Event number')['Time'].apply(lambda x: x.max() - x.min())
time_diff = df.groupby('Event number')['Time'].apply(lambda x: x.max() - x.min())
event_count = df.groupby('Event number')['Pulses in event'].count()
total_amplitude=df.groupby('Event number')['Amplitude'].sum()

fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(15, 5))

ax[0].scatter(time_diff, total_amplitude)
ax[0].set_ylabel('Total amplitude in event')
ax[0].set_xlabel('Time difference')

ax[1].scatter(time_diff, event_count)
ax[1].set_ylabel('Number of pulses in event')
ax[1].set_xlabel('Time difference')

ax[2].scatter(event_count, total_amplitude)
ax[2].set_ylabel('Total amplitude in event')
ax[2].set_xlabel('Number of pulses in event')

plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

data_for_cluster = pd.DataFrame({'event_count': event_count, 'time_diff': time_diff, 'total_amplitude': total_amplitude})
data_for_cluster = StandardScaler().fit_transform(data_for_cluster)

init_points=[[0,-2,0],[3,1.5,2], [5,1.5,5]]
kmeans = KMeans(n_clusters=3, init=init_points, random_state=42).fit(data_for_cluster)

fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(15, 5))

ax[0].scatter(data_for_cluster[:,1], data_for_cluster[:,2], c=kmeans.labels_)
ax[0].set_ylabel('Total amplitude in event')
ax[0].set_xlabel('Time difference')

ax[1].scatter(data_for_cluster[:,1], data_for_cluster[:,0], c=kmeans.labels_)
ax[1].set_ylabel('Number of pulses in event')
ax[1].set_xlabel('Time difference')

ax[2].scatter(data_for_cluster[:,0], data_for_cluster[:,2], c=kmeans.labels_)
ax[2].set_ylabel('Total amplitude in event')
ax[2].set_xlabel('Number of pulses in event')

plt.show()
```

```python
preds = df.groupby('Event number')['Event number'].count()
preds = pd.DataFrame(preds)
preds.drop(columns=['Event number'], inplace=True)
preds.reset_index(inplace = True)
preds['pred'] = pd.DataFrame(kmeans.labels_)
```

```python
data_for_cluster = pd.DataFrame({'event_count': event_count, 'time_diff': time_diff, 'total_amplitude': total_amplitude})
data_for_cluster = data_for_cluster.join(preds.set_index('Event number'), on = 'Event number')
data_for_cluster['Mean amplitude'] = data_for_cluster['total_amplitude'] / data_for_cluster['event_count']
```

```python
laser_mean_amplitude = data_for_cluster[data_for_cluster['pred'] == 0]['Mean amplitude'].mean()

counts, bins, _ = plt.hist(data_for_cluster[data_for_cluster['pred'] == 0]['Mean amplitude'], bins=np.linspace(0, 5, 80), edgecolor='black')
max_count_index = np.argmax(counts)
laser_peak_amplitude = (bins[max_count_index] + bins[max_count_index + 1]) / 2

laser_energy = 4.1 * 10**(-15) / (480 * 10**(-9) * 4.57 * 10**(-9))
coef = laser_energy / laser_peak_amplitude
data_for_cluster['Mean energy'] = data_for_cluster['Mean amplitude'] * coef

print(laser_mean_amplitude, laser_peak_amplitude)
```
